# Problem 3.1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cereal.csv', sep=';')
```

```python
df.head()
```

```python
df.drop(df.index[0], inplace = True)
```

```python
df.head()
```

```python
X = df[['name', 'mfr', 'type', 'calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight', 'cups']]
y = df['rating']
```

```python
X
```

```python
y
```

```python
X = X.join(pd.get_dummies(X['name']))
X = X.join(pd.get_dummies(X['mfr']))
X = X.join(pd.get_dummies(X['type']))
```

```python
X = X.drop(['name', 'mfr', 'type'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 15)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print("Intercept: ", model.intercept_)
print("Coefficients:")
list(zip(X, model.coef_))
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', rmse)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
model.score(X, y)
```
